```julia
using HRGVisualization
```

```julia
he1 = HyperEdge("S", ["1"], ["2"])
he2 = HyperEdge("X", ["1", "3"], ["2", "4"])
he3 = HyperEdge("Y", ["2", "4"], ["5"])
he4 = HyperEdge("Z", ["5"], ["6", "7", "8"])
```

```julia
display("image/svg+xml",he1)
```

```julia
display("image/png",he1)
```

```julia
hg = HyperGraph{String}(HyperEdge[he1, he2, he3, he4])
```

```julia
display("image/svg+xml",hg)
```

```julia
display("image/png",hg)
```

```julia
rules = Dict{String, HyperGraph{String}}(
    "S" => HyperGraph([HyperEdge("JOHN", ["_"], ["_"])]),
    "JOHN" => HyperGraph([HyperEdge("John", ["_"], ["3"]), HyperEdge("LOVES",["3"], ["_"])]),
    "LOVES" => HyperGraph([HyperEdge("loves", ["_"], ["4"]), HyperEdge("MARY", ["4"], ["_"])]),
    "MARY" => HyperGraph([HyperEdge("Mary", ["_"], ["_"])])
)
```

```julia
display("image/svg+xml",rules)
```

```julia
display("image/png",rules)
```
